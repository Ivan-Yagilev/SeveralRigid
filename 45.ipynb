## Стержень, лежащий на нескольких жёстких опорах

На рисунке выше изображен двухпролётный стержень. Пусть в левом пролёте длины $l_1$ жёсткость стержня равна $E_1I_1$, а в правом длины $l_2$ равна $E_2I_2$, сжимающие силы $P_1$ и $P_2$ различны.

Для рассмотрения устойчивости воспользуемся ранее известными формулами и распишем их для каждого пролёта:

$$v = \frac{M_0}{EIk^2}(\frac{\sin{kx}}{\sin{kl}} - \frac{x}{l})$$

Граничные условия: $v = 0; x = 0; x = l$

Так как влияние соседнего пролёта выражается в действии опорного момента $M_0$, угол поворота слева от опоры:

$$\theta_{B,1} = \frac{M_0l_1}{E_1I_1}\chi(u_1)$$

Угол поворота справа от опоры:

$$\theta_{B,2} = \frac{M_0l_2}{E_2I_2}\chi(u_2)$$

Где:

$$u_1=l_1\sqrt{\frac{P_1}{E_1I_1}}, u_2=l_2\sqrt{\frac{P_2}{E_2I_2}}$$

Углы $\theta_{B,1}$ и $\theta_{B,2}$ должны быть равны по величине и обратны по знаку; следовательно при $M_0 \neq 0$ это условие запишем в виде:

$$\frac{l_1}{E_1I_1}\chi(u_1) + \frac{l_2}{E_2I_2}\chi(u_2) = 0 \ \ \ \ \ \ (1) $$

Из уравнения (1) можно определить критическое сочетание нагрузок $P_1$ и $P_2$. Пусть жёсткости пролётов одинаковы, тогда по формуле (1):

$$\frac{\chi(u_1)}{\chi(u_2)} = - \frac{l_2}{l_1} \ \ \ \ \ \ (2)$$

Если же длины пролётов равны, то условие (1) не может быть выполнено; при этом надо положить $M_0 = 0$. Отсюда $\chi(u_1) = \chi(u_2) = \inf$; но тогда будет $u = \pi$ и критическая нагрузка определится по обычной формуле Эйлера. Уравнение (2) можно истолковать так, что в более коротком пролёте опорный момент "помогает" сжимающим силам $P$ изогнуть стержень, а в более длинном - "препятствует им".

Если допустить, что в одном из пролётов (в левом) сила $P$ является сжимающей, а в другом (в правом) - растягивающей, то вместо (1) получим:

$$ \frac{l_1}{E_1I_1}\chi(u_1) + \frac{l_2}{E_2I_2}\chi_1(u_2) = 0 \ \ \ \ \ \ (3)$$

Уравнения (1) или (3) решаются с помощью таблицы, полученной экспериментальным путём.

Пусть число пролётов больше двух. Рассмотрим пролёты $n-1, n$, выпишем условие $\theta_{n,1} + \theta_{n,2} = 0 $; но при этом учтём влияние второго опорного момента. Для трёх соседних опор применим индексы $n-1, n, n+1$.

Слева от средней опоры угол $\theta_{n,1}$:

$$\theta_{n,1} = \frac{M_{n-1}l_{n-1}}{E_{n-1}I_{n-1}}\phi(u_{n-1}) + \frac{M_{n}l_{n-1}}{E_{n-1}I_{n-1}}\chi(u_{n-1}) $$

Справа от средней опоры:
$$\theta_{n,2} = \frac{M_{n+1}l_{n}}{E_{n}I_{n}}\phi(u_{n}) + \frac{M_{n}l_{n}}{E_{n}I_{n}}\chi(u_{n}) $$

Приравнивая к нулю сумму этих выражений, получаем уравнение трёх моментов:

$$\frac{M_{n-1}l_{n-1}}{E_{n-1}I_{n-1}}\phi(u_{n-1}) + M_{n}[\frac{l_{n-1}}{E_{n-1}I_{n-1}}\chi(u_{n-1}) + \frac{l_{n}}{E_{n}I_{n}}\chi(u_{n})] + \frac{M_{n+1}l_{n}}{E_{n}I_{n}}\phi(u_{n}) = 0$$

Выписывая подобные уравнения для каждой пары пролётов, получаем систему однородных трёхчленных линейных уравнений относительно моментов $M_1, M_2$ и т.д. Система критических нагрузок будет найдена, если приравнять нулю определитель, составленный из коэффициентов при $M_1, M_2$ и т.д. в этих уравнениях.

Пусть крайние опоры являются шарнирными, введём обозначение $s = \frac{l}{EI}$, найдём окончательное уравнение в виде:

$
\begin{vmatrix}
s_0\chi(u_0) + s_1\chi(u_1) & s_1\phi(u_1) & 0 & 0 & . \\
s_1\phi(u_1) & s_1\chi(u_1) + s_2\chi(u_2) & s_2\phi(u_2) & 0 & . \\
0 & s_2\phi(u_2) & s_2\chi(u_2) + s_3\chi(u_3) & s_3\phi(u_3) & . \\
0 & 0 & s_3\phi(u_3) & s_3\chi(u_3) + s_4\chi(u_4) & . \\
. & . & . & . & .
\end{vmatrix} = 0
$

Для двухпролётного стержня определитель будет состоять только из одного члена, для трёхпролётного - два первых члена в первой и второй строке и т.д.

Расчёт стержня можно проводить по формулам (1) или (3) и в том случае, если во всех пролётах или в некоторых из них критическое напряжение превышает предел упругости. В формулы $u_{n}$ и последующие тогда следует ввести вместо модулей $E$ результирующие модули $T$ или касательные $E_к$. Надо иметь в виду, что величины этих модулей зависят от искомых критических напряжений; поэтому расчёт может быть осуществлён лишь после нескольких попыток.

### Пример

Ходовой винт $AB$ токарно-винторезного станка воспринимает усилие подачи, передающееся ему от режущего инструмента через перемещающуюся вдоль винта гайку $C$. Винт должен быть расчитан на устойчивость в предположении, что гайка находится в крайнем правом положении, на расстоянии $l_2$ от опоры $C$, и передаёт участку винта $AC$, имеющему длину $l_1$, сжимающую силу $P$.

Определить внутренний диаметр $d$ винта в двух случаях:

1. Гайка имеет малую длину и должна рассматриваться как промежуточная шарнирная опора (а);

2. Длина гайки велика, и винт следует считать защемлённым в опоре $С$ (б).

Концы винта $A, B$ в обоих случаях принимаются закреплёнными шарнирно. $P = 900 \ кг$, коэффициент запаса устойчивости $n = 3.5, l_1 = 150 \ см, l_2 = 50 \ см$. Материал - сталь, $E = 2.1 \cdot 10^6 \ кг \cdot см^2, \sigma_{пц} = 2800 \ \frac{кг}{см^2}$. При вычислении момента инерции сечения влиянием нарезки пренебречь.

**Решение:**

1. Получаем стержень с тремя опорами, усилие в первом участке равно $P_{кр}$, во втором равно нулю. Так как жесткость по длине постоянна, воспользуемся уравнением (2):

    $$\frac{\chi(u_1)}{\chi(0)} = -\frac{l_2}{l_1}$$

    В таблице находим $\chi(0) = 0.333.$ Следовательно,

    $\chi(u_1) = -\frac{50}{150} \cdot 0.333 = -0.111$

    Этому значению $\chi$ отвечает $u_1 \approx 4.1$. Тогда по формуле:

    $u = kl = l \sqrt{\frac{P}{EI}}$

    $(k_1l_1)^2 = u_1^2 = 16.81$

    Критическая сила должна быть равна $P_{кр} = nP = 3150$. Момент инерции сечения будет

    $$I = \frac{P_{кр}l_1^2}{16.81E} = \frac{3150 \cdot 150^2}{16.81 \cdot 2.1 \cdot 10^6} \approx 2 \ см^4$$

    Отсюда диаметр винта $\textbf{d = 26 мм}$, площадь сечения $5.3 \ см^2$. Критическое напряжение

    $$\sigma_{кр} = \frac{3150}{5.3} = 600 \ \frac{кг}{см^2}$$

    лежит в пределах пропорциональности.

    Если не учитывать влияние участка $CB$ и считать обе опоры $A, C$ шарнирными, то получим параметр $u_1^2$ равным $\pi^2$; критическая нагрузка оказалась бы на $41%$ меньше.

2. Винт на участке $AC$ надо рассматривать как стержень, один конец которого закреплён шарнирно, а другой защемлён. Параметр $u_1^2$ равным $2\pi^2$. Находим: $I = 1.7 \ см^4, \textbf{d = 24 мм}$
